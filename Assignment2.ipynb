**Problem 1: Write a program to calculate payments based on hour rate**

```python
def calculate_payment(hours, rate):
    """Calculate payment based on hours worked and hourly rate."""
    return hours * rate

def main():
    try:
        # Get user input
        hours = float(input("Enter the number of hours worked: "))
        rate = float(input("Enter your hourly rate: "))

        # Calculate and display payment
        payment = calculate_payment(hours, rate)
        print(f"Total payment: Rs {payment:.2f}")

    except ValueError:
        print("Please enter valid numbers for hours and rate.")
main()
```

**Problem 2: Write a program to calculate compound interest based on user input on principle, years (duration) and interest rate.**

```python
def calculate_compound_interest(principal, rate, years):
    """Calculate compound interest."""
    amount = principal * (1 + rate / 100) ** years
    compound_interest = amount - principal
    return compound_interest, amount

def main():
    try:
        # Get user input
        principal = float(input("Enter the principal amount: "))
        rate = float(input("Enter the annual interest rate (in %): "))
        years = int(input("Enter the number of years: "))

        # Calculate compound interest
        ci, total_amount = calculate_compound_interest(principal, rate, years)
        print(f"\nCompound Interest: Rs {ci:.2f}")
        print(f"Total Amount after {years} years:Rs {total_amount:.2f}")

    except ValueError:
        print("Please enter valid numeric values for principal, rate, and years.")

if __name__ == "__main__":
    main()
```
