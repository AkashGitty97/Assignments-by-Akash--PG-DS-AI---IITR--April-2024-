Q1: Coin Toss: Create a function that return 0, 1 with equal probability.


```python
from random import random
def coint_toss_fair():
    r = random()
    if r < 0.5:
        return 0
    else:
        return 1

result = [coint_toss_fair() for i in range(10000)]
print("Example for fair two sided coin toss")
print(f"Number of 0's occured : {100 * (result.count(0) / len(result)) :.2f}")
print(f"Number of 1's occured : {100 * (result.count(1) / len(result)) :.2f}")
```

Q2: Coin Toss: Create a function that return 0, 1, 2 with equal probability.

```python
from random import random

# Method 1: Fair 3-sided coin toss using digits from random float
def coin_toss_three_fair():
    rr = random()
    # Extract more digits to improve randomness
    i = int(rr * (10 ** 10))  # More precision
    return i % 3

# Method 2: Retry if random()*10 gives 0 (ensuring uniformity)
def coin_toss_three_fair2():
    i = int(random() * 10)
    if i != 0:
        return i % 3
    else:
        return coin_toss_three_fair2()  # Must return the recursive result

# Run simulations
result1 = [coin_toss_three_fair() for _ in range(10000)]
print("Example for fair three-sided coin toss, sample one")
print(f"Number of 0's occurred : {100 * (result1.count(0) / len(result1)) :.2f}%")
print(f"Number of 1's occurred : {100 * (result1.count(1) / len(result1)) :.2f}%")
print(f"Number of 2's occurred : {100 * (result1.count(2) / len(result1)) :.2f}%\n")

result2 = [coin_toss_three_fair2() for _ in range(10000)]
print("Example for fair three-sided coin toss, sample two")
print(f"Number of 0's occurred : {100 * (result2.count(0) / len(result2)) :.2f}%")
print(f"Number of 1's occurred : {100 * (result2.count(1) / len(result2)) :.2f}%")
print(f"Number of 2's occurred : {100 * (result2.count(2) / len(result2)) :.2f}%")
```

Q3: Create an faced die which generates number from 0 to n - 1 with equal probability.

```python
from random import random
def coint_toss_fair(n):
    rr = random()
    ii = int(rr*(10**(len(str(rr))-2)))
    return (ii % n)

num = 10
result = [coint_toss_fair(num) for i in range(10000)]
print("Example for fair {num} sided coin toss")

total=0
for i in range(num):
    print(f"Number of {i}'s occured : {100 * (result.count(i) / len(result)) :.2f}")
    total+=result.count(i) / len(result)
    i+=1

print(f"Total : {total :.2f}")
```
