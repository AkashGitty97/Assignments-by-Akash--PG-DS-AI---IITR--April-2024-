Problem 1. Print all strings that can be generated from a list of letters.
Example:
Input: abc
Output:
abc
acb
bac
bca
cab
cba

```python
def generate_permutations(s, answer="", seen=None):
    if seen is None:
        seen = set()

    if len(s) == 0:
        if answer not in seen:
            print(answer)
            seen.add(answer)
        return

    for i in range(len(s)):
        ch = s[i]
        left = s[:i]
        right = s[i+1:]
        rest = left + right
        generate_permutations(rest, answer + ch, seen)

def main():
    user_input = input("Enter a string of letters: ")
    print("\nUnique permutations:")
    generate_permutations(user_input)

if __name__ == "__main__":
    main()
```

The above code ensures that each unique permutation is printed only once.

**Problem 2:Code Huber Loss**

```python
def huber_loss(y_true, y_pred, delta=1.0):
    loss = 0
    for yt, yp in zip(y_true, y_pred):
        error = yt - yp
        if abs(error) <= delta:
            loss += 0.5 * error ** 2
        else:
            loss += delta * (abs(error) - 0.5 * delta)
    return loss / len(y_true)

# Get user input
y_true = list(map(float, input("Enter actual values (space-separated): ").split()))
y_pred = list(map(float, input("Enter predicted values (space-separated): ").split()))
delta = float(input("Enter delta (e.g. 1.0): "))

# Calculate and print Huber Loss
if len(y_true) == len(y_pred):
    loss = huber_loss(y_true, y_pred, delta)
    print(f"Huber Loss: {loss:.4f}")
else:
    print("Error: y_true and y_pred must be of the same length.")
```

**Problem 3:Code for BLEU Score**

```python
def bleu_score(reference, candidate):
    reference_words = reference.split()
    candidate_words = candidate.split()

    matches = 0
    for word in candidate_words:
        if word in reference_words:
            matches += 1
            reference_words.remove(word)  # prevent counting the same word again

    precision = matches / len(candidate_words) if candidate_words else 0
    return precision

# User input
reference = input("Enter reference sentence: ").lower()
candidate = input("Enter candidate sentence: ").lower()

score = bleu_score(reference, candidate)
print(f"\nBLEU-1 Score: {score:.4f}")
```

**Problem 4:Design for implementing multiplication without using the multiplication operator**

```python
def multiply(a, b):
    result = 0
    negative = False

    # Handle negatives
    if a < 0:
        a = -a
        negative = not negative
    if b < 0:
        b = -b
        negative = not negative

    for _ in range(b):
        result += a

    return -result if negative else result

# Get input from user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Perform multiplication
product = multiply(a, b)
print(f"\nResult: {a} x {b} = {product}")
```
