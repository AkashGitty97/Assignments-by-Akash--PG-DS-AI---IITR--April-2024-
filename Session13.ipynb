#Find the next n lines line in the file (line ending with /n)


```python
def read_n_lines(filename,n):
    count = 0
    lines = []
    f = open(filename)
    while (count < n):
        ln = f.readline()
        if len(ln) > 0:
            lines.append(ln)
        else:
            break
        count+=1
    f.close()
    return lines
```

Find the largest line in the file (line ending with /n)

```python
def find_longest_line(filename):
    f = open(filename)
    txtlines = f.readlines()
    longest_linenumber = 0
    for i in range(len(txtlines)):
        if (len(txtlines[longest_linenumber]) < len(txtlines[i])):
            longest_linenumber = i
    f.close()
    return txtlines[longest_linenumber]

longest_line = find_longest_line('/cxldata/big.txt')
print(longest_line)

```
