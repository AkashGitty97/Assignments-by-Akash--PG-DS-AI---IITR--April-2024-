**Assignment X.X : Predict the next number**

A linear relationship is assumed for input-output pairs.:

y = m * x + b
We rearrange it to solve for x:
x = (y - b) / m

```python
import matplotlib.pyplot as plt

def predict_x_from_y(x_values, y_values, target_y):
    n = len(x_values)
    sum_x = sum(x_values)
    sum_y = sum(y_values)
    sum_xy = sum(x * y for x, y in zip(x_values, y_values))
    sum_x2 = sum(x ** 2 for x in x_values)

    denominator = n * sum_x2 - sum_x ** 2
    m = (n * sum_xy - sum_x * sum_y) / denominator
    b = (sum_y - m * sum_x) / n

    predicted_x = (target_y - b) / m
    return predicted_x, m, b

def plot_relationship(x_vals, y_vals, m, b, predicted_x, target_y):
    plt.scatter(x_vals, y_vals, color='blue', label='Original Data')

    # Regression line
    x_line = list(range(int(min(x_vals)) - 1, int(max(x_vals)) + 2))
    y_line = [m * x + b for x in x_line]
    plt.plot(x_line, y_line, color='green', label='Regression Line')

    # Predicted point
    plt.scatter([predicted_x], [target_y], color='red', marker='x', s=100, label='Predicted Point')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('X-Y Relationship and Predicted Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# --- Main ---
if __name__ == "__main__":
    x_vals = list(map(float, input("Enter x values (space-separated): ").split()))
    y_vals = list(map(float, input("Enter y values (space-separated): ").split()))
    y_input = float(input("Enter a value of y to predict x: "))

    predicted_x, m, b = predict_x_from_y(x_vals, y_vals, y_input)
    print(f"\nPredicted x for y = {y_input} is: {predicted_x:.4f}")

    plot_relationship(x_vals, y_vals, m, b, predicted_x, y_input)
```
